```python
import cv2

# resize images - this way the NN can process them faster 
# we tried different sizes and 64x64 seems small and detailed enough
for file in f:
    # read image
    img = cv2.imread('pokemon/' + file)
    # resize image
    img = cv2.resize(img, dsize=(64, 64), interpolation=cv2.INTER_CUBIC)
    # save image
    cv2.imwrite('pokemon_small/' + file, img)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from os import walk

# this array will contain the RGB images
imgs = []

f = []
for (dirpath, dirnames, filenames) in walk('pokemon'):
    f.extend(filenames)

for src in f:
    # read PNG images and convert them to RGB tensors
    img = mpimg.imread('pokemon_small/'+src)[:,:,0:3]
    # add image to the list
    imgs.append(img)
    
imgs = np.array(imgs)
    
print('Number of images: %d\nImage size: (%d, %d)' % (imgs.shape[0], imgs.shape[1], imgs.shape[2]))
    
#plot the images
fig, axs = plt.subplots(5, 5, figsize = (10,10))
for i in range(5):
    for j in range(5):
        axs[i,j].imshow(imgs[i*5+j])        
```
